## When to use a 'groupby' in pandas

```python
drinks = pd.read_csv('videos_data/drinks.csv')
```

```python
drinks.head()
```

**Question: what is the average beer serving by continent**

```python
#grouping is multiple fitering of (a or some) category
drinks.groupby('continent').beer_servings.mean() 
```

```python
# filtering is extracting an insight by specific condition
drinks.loc[drinks['continent']=='Africa','beer_servings'].mean()
```

```python
drinks.groupby('continent').beer_servings.max()
```

## When to use the `inplace` parameter in pandas

```python
ufo = pd.read_csv('videos_data/ufo.csv')
```

```python
ufo.head()
```

```python
ufo.drop('City',axis=1).head() 
```

```python
ufo.head()
```

```python
ufo.drop('City',axis=1,inplace=True)
```

```python
ufo.head()
```

```python
ufo.dropna(how='any')
```

```python
ufo.dropna(how='any').shape
```

```python
ufo.shape
```

```python
# assignment operator as an alternative to inplace parameter
ufo = ufo.set_index('Time')
```

```python
ufo.tail()
```

```python
# inplace is use for experimenting
ufo.fillna(method='bfill').tail()
```

```python
ufo.fillna(method='ffill').tail()
```
