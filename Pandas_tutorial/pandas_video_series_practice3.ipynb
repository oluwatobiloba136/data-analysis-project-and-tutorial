## The use of `map`, `apply` and `applymap`

```python
train = pd.read_csv('videos_data/titanic_train.csv')
```

```python
train.head(3)
```

```python
# map is a series method
train['sex_male'] = train.Sex.map({'male':1,'female':0})
```

```python
train.loc[0:4,['Sex','sex_male']]
```

```python
# apply is a series and dataframe function
train['name_length'] = train.Name.apply(len)
```

```python
train.loc[0:4,['Name','name_length']]
```

```python
train['fare_rounded'] = train.Fare.apply(np.ceil)
```

```python
train.loc[0:4,['Fare','fare_rounded']]
```

```python
train.Name.str.split(',').apply(lambda x:x[0]).head()
```

```python
drinks.head(2)
```

```python
# apply as a dataframe function
drinks.loc[:,'beer_servings':'wine_servings'].apply(max,axis=0)
```

```python
drinks.loc[:,'beer_servings':'wine_servings'].apply(max,axis=1)
```

```python
drinks.loc[:,'beer_servings':'wine_servings'].apply(np.argmax,axis=0)
```

```python
# applymap is used for every element in the dataframe
drinks.loc[:,'beer_servings':'wine_servings'].applymap(float)
```

```python
drinks.loc[:,'beer_servings':'wine_servings'] = drinks.loc[:,'beer_servings':'wine_servings'].applymap(float)
```

## How to extract year, month[padded]... as string from datetime object

```python
ufo = pd.read_csv('videos_data/ufo.csv')
```

```python
ufo['Time'] = pd.to_datetime(ufo.Time)
```

```python
ufo['year'] = ufo.Time.dt.year.astype(str)
```

```python
ufo['month'] = ufo.Time.dt.month.astype(str).str.pad(width=2, fillchar='0')
```

```python
ufo.head(3)
```

```python
ufo.dtypes
```

## How to concatenate two or more series

```python
ufo['new']= ufo.year.str.cat(ufo.month, sep='-')
```

```python
ufo.head(3)
```

## How to do multi-conditional mapping

```python
train = pd.read_csv('videos_data/titanic_train.csv')
```

```python
train.head(3)
```

```python
train['young_male'] = ((train.Sex == 'male')&(train.Age < 30)).map({True:'yes',False:'no'})
```
